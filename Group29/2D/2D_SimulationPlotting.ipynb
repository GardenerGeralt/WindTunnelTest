```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv('NACA 642A015_T1_Re0.700_M0.14_N9.0.txt', header=0, sep=' ', skipinitialspace=True)
```

```python
data.head()
```

```python
data.drop(index=0, inplace=True)
data.head()
```

```python
data.columns
```

```python
#CL plot overview

x = data['alpha'].values.astype(np.float)
y = data['CL'].values.astype(np.float)

plt.style.use('seaborn-whitegrid')
plt.rcParams["axes.edgecolor"] = "0.15"
plt.rcParams["axes.linewidth"]  = 1.25

plt.plot(x, y, linestyle='solid', marker='o', markersize=4)
plt.ylabel("Lift coefficient [-]", fontsize=16)
plt.xlabel("Angle of attack [deg]", fontsize=16)
# plt.tight_layout(pad=1, h_pad=0.5, w_pad=0.5)
# plt.legend()
plt.savefig("CL2D_simulation.pdf")
```

```python
#CD plot overview

x = data['alpha'].values.astype(np.float)
y = data['CD'].values.astype(np.float)

plt.style.use('seaborn-whitegrid')
plt.rcParams["axes.edgecolor"] = "0.15"
plt.rcParams["axes.linewidth"]  = 1.25

plt.plot(x, y, linestyle='solid', marker='o', markersize=4)
plt.ylabel("Drag coefficient [-]", fontsize=16)
plt.xlabel("Angle of attack [deg]", fontsize=16)
# plt.tight_layout(pad=1, h_pad=0.5, w_pad=0.5)
# plt.legend()
plt.savefig("CD2D_simulation.pdf")
```
